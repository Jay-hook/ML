```python
import numpy as np
import pandas as pd
# Datetimeindex
dates= pd.date_range('2020-01-01', periods=48, freq='M')

# additive mode:trend + cycle + seasonality + irregular factor
timestamp = np.arange(len(dates))
trend_factor = timestamp*1.1
cycle_factor = 10*np.sin(np.linspace(0, np.pi*2, 48))
seasonal_factor = 7*np.sin(np.linspace(0, np.pi*8, 48))
np.random.seed(2004)
irregular_factor = 2*np.random.randn(len(dates))

df = pd.DataFrame({'timeseries':trend_factor + cycle_factor + seasonal_factor + irregular_factor,
                  'trend': trend_factor,
                  'cycle': cycle_factor,
                  'seasonal': seasonal_factor,
                  'irregular': irregular_factor},
                 index=dates)                        
```

```python
# Time series plot
import matplotlib.pyplot as plt

plt.figure(figsize=[10,6])
df.timeseries.plot()
plt.title('Time Series (Additive Mpdel)', fontsize=16)
plt.ylim(-12, 55)
plt.show()
```
